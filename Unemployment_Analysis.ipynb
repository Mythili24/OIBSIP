###### Importing the libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

###### Reading the input car dataset and loading it into a DataFrame

```python
df = pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")
df
```

###### Info about the columns from the input dataset
Frequency = measuring frequency (Monthly)

Estimated Unemployment Rate (%) = percentage of people unemployed in each States of India

Estimated Employed = Number of people employed

Estimated Labour Participation Rate (%) = The labour force participation rate is the portion of the working population in the 16-64 years' age group in the economy currently in employment or seeking employment.

```python
df.info()
```

```python
df.columns
```

###### Renaming the columns

```python
df.rename(columns = {'Region':'States', ' Date':'Date', ' Frequency':'Frequency', ' Estimated Unemployment Rate (%)':'Estimated_Unemployment_Rate',
                                        ' Estimated Employed':'Estimated_Employed', ' Estimated Labour Participation Rate (%)':'Estimated_Labour_Participation_Rate',
                                       'Region.1':'Region', 'longitude':'Longitude', 'latitude':'Latitude'}, inplace = True)
df
```

```python
df.info()
```

###### Changing the data types of categorical columns

```python
df['States'] = df['States'].astype('string')
df['Date'] = pd.to_datetime(df['Date'])
df['Frequency'] = df['Frequency'].astype('string')
df['Region'] = df['Region'].astype('string')
```

```python
df.info()
```

###### Checking for null values in the DataFrame

```python
df.isnull().sum()
```

###### Observation: There are no null values in the dataset

###### Deriving the statistical summary of DataFrame

```python
df.describe()
```

###### Dropping the string and date columns

```python
df_num = df.select_dtypes(exclude=['string', 'datetime64[ns]'])
df_num
```

###### Unpivoting a DataFrame from wide format to long format.

```python
df_melted = pd.melt(df_num)
df_melted
```

```python
df_melted.info()
```

###### Creating a box plot from melted Dataframe

```python
plt.figure(figsize=(20,15))
sns.boxplot(x="variable", y="value", data=df_melted)
plt.show()
```

###### Observation: We can infer from above box plot that only Estimated_Employed column has outliers

###### Creating a Pair Plot

```python
plt.figure(figsize=(20,10))
sns.pairplot(df_num)
plt.show()
```

###### Creating correlation matrix from the DataFrame

```python
cor = df_num.corr()
cor
```

###### Plotting the correlation matrix onto a Heatmap

```python
# figure size
plt.figure(figsize=(16,8))

# heatmap
#colormap = sns.color_palette("Greens")
sns.heatmap(cor, cmap="PiYG", annot=True)
plt.show()
```

###### Observations: Both Estimated_Employed and Estimated_Labour_participation_rate are negatively related to Estimated_Unemployment_Rate 

```python
df.columns
```

###### Finding the mean of Estimated_Unemployment_Rate Statewise and sorting them to get Top 10 States with their mean values

```python
EUR_mean = df.groupby(['States'])['Estimated_Unemployment_Rate'].mean().reset_index()
EUR_mean.sort_values('Estimated_Unemployment_Rate', ascending=False, inplace=True)
EUR = EUR_mean.head(10)
EUR
```

###### Plotting the Top 10 States with their Mean Estimated Unemployment Rates

```python
plt.figure(figsize=(15,10))
sns.barplot(x=EUR['States'], y=EUR['Estimated_Unemployment_Rate'], color="gold")
plt.ylabel('Estimated_Unemployment_Rate', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.title("Top 10 States with their Mean Estimated_Unemployment_Rate", fontsize=15)
plt.show()
```

##### Observation : Haryana has highest mean Estimated_Unemployment_Rate at 27.477000

###### Plotting the trend of Estimated Unemployment Rate from Feb 2020 to Nov 2020 for each region

```python
plt.figure(figsize=(15,10))
sns.lineplot(x="Date", y="Estimated_Unemployment_Rate", data=df, hue="Region")
 
plt.show()
```

###### Observation : We can infer from the above graph that for every region EUR has reached maximum during Apr 2020 to July 2020 i.e. during first wave of Covid-19

###### Plotting Sunburst Chart to display Unemployment rate in a hierarchical manner (Region and State)

```python
import plotly.express as px
sun_df = df[['States', 'Region', 'Estimated_Unemployment_Rate']]
fig = px.sunburst(sun_df, path=['Region', 'States'], values='Estimated_Unemployment_Rate',
                  title= 'Unemployment rate in every State and Region', height=650)
fig.show()
```

###### Observations: Sunburst Chart displays EUR at both Region and State level. North region contribution is more than others.

###### Comparing EUR and ELPR for each region using multiple bar charts

```python
region_mean = df.groupby(['Region'])['Estimated_Unemployment_Rate', 'Estimated_Labour_Participation_Rate'].mean().reset_index()
region_mean.sort_values('Estimated_Unemployment_Rate', ascending=False, inplace=True)
regionmean = region_mean.head(10)
regionmean

Y = regionmean['Estimated_Unemployment_Rate'].tolist()
Z = regionmean['Estimated_Labour_Participation_Rate'].tolist()

N=5
ind = np.arange(N)
width = 0.25

plt.bar(ind, Y, width=width, color='g', label='Estimated_Unemployment_Rate')


plt.bar(ind + width, Z, width=width, color='b', label='Estimated_Labour_Participation_Rate')
 

plt.xlabel("Region")
plt.ylabel("Unemployment vs Labour Participation")
plt.title("Unemployment vs Labour Participation in each region")

# Adding Xticks
plt.xticks(ind+width/2, ['North', 'East', 'Northeast', 'South', 'West'])
plt.legend()
plt.show()
```

###### Observation : 1) Maximum Labour had participated in NorthEast Region. 
######                         2) Maximum unemployment is there is North Region

###### Plotting the mean Estimated Employment contribution for each region in a Pie Chart

```python
EE_mean = df.groupby(['Region'])['Estimated_Employed'].mean().reset_index()
EE_mean.sort_values('Estimated_Employed', ascending=False, inplace=True)
EE_mean
```

```python
data = EE_mean['Estimated_Employed'].tolist()
keys = ['North', 'East', 'Northeast', 'South', 'West']

# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0)
 
# Creating color parameters
colors = ( "orange", "cyan", "brown", "grey", "indigo")
# plotting data on chart
plt.pie(data, labels=keys, colors=colors,
        explode=explode, autopct='%.0f%%')
plt.title('Regionwise Mean Estimated Employed contribution')
  
# displaying chart
plt.show()
```

###### Observation: North region has maximum contribution of 28% and West region has the minimum contribution of 5%.

###### Plotting Top 10 States with their mean Estimated Employment in Bar graph

```python
employed_mean = df.groupby(['States'])['Estimated_Employed'].mean().reset_index()
employed_mean.sort_values('Estimated_Employed', ascending=False, inplace=True)
employed_states = employed_mean.head(10)
employed_states
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=employed_states['States'], y=employed_states['Estimated_Employed'], color='cyan')
plt.ylabel('Mean_Estimated_Employed', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.title("Top 10 States with their mean Estimated_Employed", fontsize=15)
plt.show()
```
